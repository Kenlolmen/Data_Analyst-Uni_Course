```python
#Zadanie4

readcsv = pd.read_csv('autos.csv')

avgFuel = readcsv.groupby('make')[['city-mpg', 'highway-mpg']].mean()

print(avgFuel)
```

```python
#zadanie 5

CountFuel = readcsv.groupby('make')['fuel-type'].value_counts().unstack()
print(CountFuel)
```

```python
#zadanie 6

x = readcsv['length'].values
y = readcsv['city-mpg'].values

wiel1 = np.polyfit(x,y,1)
wiel2 = np.polyfit(x,y,2)

y1 = np.polyval(wiel1,x)
y2 = np.polyval(wiel2,x)

print(wiel1)
print(wiel2)

#zadanie 7

from scipy.stats import pearsonr

cor, _  = pearsonr(x,y)
print(f'korelacja : {cor}')
```

```python
#zadanie 8 
import matplotlib.pyplot as plt

lin = np.linspace(x.min(), x.max(),100)

line1 = np.polyval(wiel1, lin)
line2 = np.polyval(wiel2, lin)

plt.scatter(x, y, color='orange', label='probki')
plt.plot(lin, line1, color='blue', label='1 stopnia ')
plt.plot(lin, line2, color='green', label='2 stopnia')
plt.legend()
plt.show()
```

```python
from scipy.stats import gaussian_kde
 
estymator_length = gaussian_kde(x)
x_range_length  = np.linspace(x.min(), x.max(), 1000)
y_estymator_length  = estymator_length(x_range_length )

plt.hist(x, bins=30, density=True, alpha=0.5, color='green', label='Histogram próbek')
plt.plot(x_range_length , y_estymator_length , color='red', label='estymator KDE')
plt.legend()
plt.show()
```

```python
#zadanie 10

z = readcsv['width'].values
estymator_width = gaussian_kde(z)
x_range_width  = np.linspace(z.min(), z.max(), 1000)
y_estymator_width = estymator_width(x_range_width)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].hist(x, bins=30, density=True, alpha=0.5, color='green', label='Histogram próbek')
ax[0].plot(x_range_length , y_estymator_length , color='red', label='estymator')
ax[0].set_title('Length')
ax[0].legend()

ax[1].hist(z, bins=30, density=True, alpha=0.5, color='blue', label='Histogram próbek')
ax[1].plot(x_range_width , y_estymator_width , color='red', label='estymator')
ax[1].set_title('Width')
ax[1].legend()
```

```python
#zadanie 11

data = np.vstack([x,z])
estymator = gaussian_kde(data)

x_grid, y_grid = np.meshgrid(np.linspace(x.min(), x.max(), 100), np.linspace(z.min(), z.max(), 100))
grid_points = np.vstack([(x_grid.ravel(), y_grid.ravel())])
val = estymator(grid_points).reshape(x_grid.shape)

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, z, 'go', label = 'probki danych')
ax.contour(x_grid, y_grid, val, cmap = 'Blues')
ax.legend()

plt.savefig("estymator_gestosci.png")
plt.savefig("estymator_gestosci.pdf")
```
